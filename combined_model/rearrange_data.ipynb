```python
data = np.load(data_path)
```

```python
import os
folder_name = '/home/shiqi/code/Project2-sensor-case/model_combination_Argos/combined_model_20240805/outputs/experiment_1/hybrid_traj/'
if not os.path.exists(folder_name):
    os.makedirs(folder_name)
else:
    print('folder already exists')

for i in range(len(data)):

    subfolder_name = os.path.join(folder_name, f"hybrid_traj_{i}")
    os.makedirs(subfolder_name, exist_ok=True)

    N = data[i].shape[0]
    for j in range(1, N + 1):
        vector = data[i][j - 1, :]
        time = 0.004 * j
        filename = os.path.join(subfolder_name, f"Solution_{j}.txt")
        with open(filename, 'w') as file:
            file.write(f"time = {time}\n")
            for value in vector:
                file.write(f"{value}\n")
        print(f"data has been saved to {filename}")
```
